```python
#SEABORN
#LINE PLOT

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data={"Days":[1,2,3,4,5],
     "NOP":[30,20,50,10,40]}
df=pd.DataFrame(data)
print(df)

sns.lineplot(data=data, x="Days", y="NOP")
plt.show()
```

```python
data=pd.read_excel("ESD.xlsx")
print(data)

color=sns.color_palette('viridis')
sns.lineplot(data=data, x='Business Unit', y="Age", hue='Ethnicity',palette=color) # style=
plt.show()
```

```python
#STYLE AND COLOR PLOT
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("tips.csv")
df

```

```python
sns.barplot(data=df,x="sex", y="tip")
sns.set_style(style="darkgrid")
plt.show()
```

```python
sns.barplot(data=df,x="sex", y="tip")
sns.set_style(style='ticks')
plt.show()
```

```python
#PALPLOT
sns.palplot(sns.color_palette("spring"))
plt.show()
```

```python
sns.palplot(sns.color_palette("viridis"))
plt.show()
```

```python
# MULTIPLE PLOT IN SEABORN(FACETGRID)

sns.FacetGrid(df, col="smoker",height=2)
plt.show()
```

```python
sns.FacetGrid(df, col="day")
plt.show()
```

```python
a=sns.FacetGrid(df, col="time",height=2,hue="sex")
a.map(sns.barplot, "day","tip")
plt.show()
```

```python
#RELATIONALNAL PLOT

sns.relplot(data=df,x="tip",y="total_bill",hue="sex",kind="line",col="day")
plt.show()
```

```python
sns.relplot(data=df,x="tip",y="total_bill",hue="sex",col="day")
plt.show()
```

```python
#SWARM PLOT- not overlap to each other unlike strip plot

sns.swarmplot(data=df, x="day", y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
sns.stripplot(data=df,x="day",y="total_bill",jitter=0.3,hue="sex")
plt.show()
```

```python
#KDE PLOT

sns.kdeplot(data=df,x="total_bill")
plt.show()
```
